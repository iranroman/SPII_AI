```
# loading the cancer dataset
# Do NOT change anything in this cell

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn import datasets
X, y = datasets.load_breast_cancer(return_X_y=True)

print('The shape of X is: ', X.shape)
print('The shape of y is: ', y.shape)
```

```
# select the two features you want to visualize
# only change this cell
N = X.shape[0]
X_2D = np.zeros([N,2])
X_2D[:,0] = X[:,0] # <- change this line
X_2D[:,1] = X[:,1] # <- change this line
```

```
# this cell does the visualization for you
# Do NOT change anything in this cell

colors = ['r','g']

cancer_idx = np.where(y==0)
nocancer_idx = np.where(y==1)
  
plt.scatter(X_2D[cancer_idx,0],X_2D[cancer_idx,1],c=colors[0],label='Cancer')
plt.scatter(X_2D[nocancer_idx,0],X_2D[nocancer_idx,1],c=colors[1],label='No cancer')  
plt.legend()
plt.show()
```
